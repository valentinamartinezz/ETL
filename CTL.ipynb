```python
CO2 = CO2.dropna() 
```

```python
CO2
```

```python
#En la columna CO2(kt)eliminar ',' y '.'
CO2['CO2 (kt)'] = CO2['CO2 (kt)'].str.replace(',','')
CO2['CO2 (kt)'] = CO2['CO2 (kt)'].str.replace('.','')
```

```python
CO2.head(10)
```

```python
CO2['CO2 (kt)'] = CO2['CO2 (kt)'].astype (int)
```

```python
CO2['CO2 per cápita (toneladas métricas)']=CO2['CO2 per cápita (toneladas métricas)'].str.replace(',','').str.replace('.','')
CO2['CO2 per cápita (toneladas métricas)']=CO2['CO2 per cápita (toneladas métricas)'].astype(float)
```
